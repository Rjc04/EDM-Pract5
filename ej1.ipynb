```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import PartialDependenceDisplay
from sklearn.model_selection import train_test_split
```

```python
# --- Cargar day.csv ---
bike_data = pd.read_csv("day.csv")

# --- Preprocesamiento ---
# Variables relevantes
features = ['instant', 'temp', 'hum', 'windspeed']
target = 'cnt'

X = bike_data[features]
y = bike_data[target]
```

```python
# --- Entrenar Random Forest ---
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# --- PDP 1D ---
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
PartialDependenceDisplay.from_estimator(model, X, features=[0], ax=ax[0,0])
PartialDependenceDisplay.from_estimator(model, X, features=[1], ax=ax[0,1])
PartialDependenceDisplay.from_estimator(model, X, features=[2], ax=ax[1,0])
PartialDependenceDisplay.from_estimator(model, X, features=[3], ax=ax[1,1])
plt.suptitle('Partial Dependence Plots - Bike Rentals (day.csv)', fontsize=16)
plt.tight_layout()
plt.show()
```
